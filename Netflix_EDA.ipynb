```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix Movies_TV shows.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
sns.countplot(x="type", data=df,color="orange")
plt.title("Movies vs TV Shows")
plt.show()
```

```python
df["type"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.ylabel("")
plt.title("Movies vs TV Shows Share")
plt.show()
```

```python
year_data = df["release_year"].value_counts().sort_index()

year_data.plot(kind="line",color="red")
plt.title("Content Release Trend")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
top_countries = df["country"].value_counts().head(10)

top_countries.plot(
    kind="barh",
    color="purple"
)
plt.title("Top 10 Countries")
plt.show()
```

```python
sns.boxplot(x=df["release_year"],color="red")
plt.title("Release Year Distribution")
plt.show()
```

```python
df["rating"].value_counts().head(10).plot(
    kind="bar",
    color="green"
)

plt.title("Top Ratings")
plt.xticks(rotation=45)
plt.show()
```

```python
Key Insights:
- Movies dominate Netflix content.
- United States contributes the highest number of titles.
- Content production increased significantly after 2015.
- TV-MA is among the most common ratings.
```
